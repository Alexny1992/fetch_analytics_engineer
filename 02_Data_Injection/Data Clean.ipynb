```python
import pandas as pd
import json
import uuid
from datetime import datetime
```

```python
# read files through google drive
users = "/content/drive/MyDrive/fetch_data/users.json"
brands = "/content/drive/MyDrive/fetch_data/brands.json"
receipts = "/content/drive/MyDrive/fetch_data/receipts.json"
```

```python
# read json files into dataframes
df_users = pd.read_json(users, lines=True)
df_brands = pd.read_json(brands, lines=True)
df_receipts = pd.read_json(receipts, lines=True)
```

```python
# checking table size
print(f"Shape of users dataframe: ", df_users.shape)
print(f"Shape of brands dataframe: ", df_brands.shape)
print(f"Shape of receipts dataframe: ", df_receipts.shape)
```

# df_user


```python
# clean up df_users
df_users['_id'] = df_users['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else None)
df_users['createdDate'] = df_users['createdDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)
df_users['lastLogin'] = df_users['lastLogin'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)

# convert date to datetime format
df_users['createdDate'] = pd.to_datetime(df_users['createdDate'], unit='ms')
df_users['lastLogin'] = pd.to_datetime(df_users['lastLogin'], unit='ms')

df_users.head()
```

```python
# write users table to csv
df_users.to_csv('df_users.csv', index=False)
```

# df_brands

```python
# extract id out of _id column
df_brands['_id'] = df_brands['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else None)
df_brands
```

```python
# writing brands to csv
df_brands.to_csv('df_brands.csv', index=False)
```

```python
# # cleaning up receipts dataframe
df_receipts['_id'] = df_receipts['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else None)
df_receipts['createDate'] = df_receipts['createDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)
df_receipts['dateScanned'] = df_receipts['dateScanned'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)
df_receipts['finishedDate'] = df_receipts['finishedDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)
df_receipts['modifyDate'] = df_receipts['modifyDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)
df_receipts['pointsAwardedDate'] = df_receipts['pointsAwardedDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)
df_receipts['purchaseDate'] = df_receipts['purchaseDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else None)

# convert date format to time
df_receipts['createDate'] = pd.to_datetime(df_receipts['createDate'], unit='ms')
df_receipts['dateScanned'] = pd.to_datetime(df_receipts['dateScanned'], unit='ms')
df_receipts['finishedDate'] = pd.to_datetime(df_receipts['finishedDate'], unit='ms')
df_receipts['modifyDate'] = pd.to_datetime(df_receipts['modifyDate'], unit='ms')
df_receipts['pointsAwardedDate'] = pd.to_datetime(df_receipts['pointsAwardedDate'], unit='ms')
df_receipts['purchaseDate'] = pd.to_datetime(df_receipts['purchaseDate'], unit='ms')

df_receipts
```

```python
# drop rewardsReceiptItemList column

df_receipts_final = df_receipts.drop(columns=['rewardsReceiptItemList'])
df_receipts_final

# write df_receipt to csv
df_receipts_final.to_csv('df_receipts.csv', index=False)
```

```python

# select the _id and rewardsReceiptItemList within receipt columns
df_receipt_data = df_receipts[['_id', 'rewardsReceiptItemList']]

df_receipt_data['_id'] = df_receipt_data['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)
# explode the rewardsReceiptItemList within the receipt column
df_exploded = df_receipt_data.explode('rewardsReceiptItemList', ignore_index=True)

# normalize rewardsReceiptItemList
df_normalized = pd.json_normalize(
    df_exploded['rewardsReceiptItemList'],
    errors='ignore'
)

# concatenate _id and normalized data
df_rewardsReceiptItemList = pd.concat([df_exploded['_id'].reset_index(drop=True), df_normalized], axis=1)

# display the rewardsReceiptItemList DataFrame
df_rewardsReceiptItemList
```

```python
# writing df_rewardsReceiptItemList to csv
df_rewardsReceiptItemList.to_csv('df_rewardsReceiptItemList.csv', index=False)
```
